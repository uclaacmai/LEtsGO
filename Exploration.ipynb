```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# load the csv files
data_path = "data/archive"
path_inds = pd.read_csv(os.path.join(data_path, "index.csv"))
class_names = pd.read_csv(os.path.join(data_path, "metadata.csv"))
```

```python
# combine the csv files
df_pre = path_inds.merge(class_names, on="class_id")
df = df_pre.drop(labels=["lego_ids", "lego_names"], axis=1)
df
```

```python
fig, axs = plt.subplots(6, 6, figsize=(18, 18))
for i, (ind, row) in enumerate(df.groupby("class_id").sample(n=1).iterrows()):
    image = plt.imread(os.path.join(data_path, row['path']))
    name = row['minifigure_name']
    axs.flat[i].imshow(image)
    axs.flat[i].set_title(name)
    axs.flat[i].axis('off')
plt.show()
```
